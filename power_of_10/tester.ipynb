```python

events = {'10K', '5K', 'Mar', 'HM', 'parkrun'}

def get_and_reshape_athlete_data(surname, club):
    # Get the list of athlete IDs
    athlete_ids = get_ids(surname=surname, club=club)
    
    # Initialize an empty list to store athlete data
    all_data = []
    
    # Loop through each athlete ID
    for athlete_id in athlete_ids:
        # Get the personal best data for the athlete
        pb_data = get_athlete_pb(athlete_id)
        
        # Add the athlete ID to each record
        for record in pb_data:
            record['athlete_id'] = athlete_id
            all_data.append(record)
    
    # Create a DataFrame from the list of all data
    df = pd.DataFrame(all_data)
    
    # Pivot the DataFrame to have events as columns
    pivot_df = df.pivot(index='athlete_id', columns='event', values='value')
    
    # Reset the index to turn the index into a column
    pivot_df.reset_index(inplace=True)
    
    # Filter the DataFrame to include only the specified events
    pivot_df = pivot_df[['athlete_id'] + list(events)]
    
    return pivot_df

# Example usage
surname = 'a'
club = 'leigh-on-sea'
reshaped_df = get_and_reshape_athlete_data(surname, club)

print(reshaped_df)
```

```python
def get_all_athlete_data(club):
    # Initialize an empty DataFrame to store all athlete data
    all_athlete_data = pd.DataFrame()
    #  defghijklmnopqrstuvwxyz
    # Loop through each letter in the alphabet
    for char in 'abc':
        # Get the reshaped athlete data for the current letter
        reshaped_df = get_and_reshape_athlete_data(surname=char, club=club)
        
        # Append the current DataFrame to the all_athlete_data DataFrame
        all_athlete_data = pd.concat([all_athlete_data, reshaped_df], ignore_index=True)
    
    return all_athlete_data

club = 'leigh-on-sea'
all_data_df = get_all_athlete_data(club)
print(all_data_df)
```
